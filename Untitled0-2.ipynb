```python

corr_mat = df.corr()
f, ax = plt.subplots(figsize=(29,15),dpi=200)
mask = np.triu(np.ones_like(corr_mat, dtype=bool))
mask = mask[1:,:-1]
corr = corr_mat.iloc[1:,:-1].copy()
sns.heatmap(corr, mask = mask,vmin=-0.6,vmax=0.7,center=0,cmap='RdPu_r',square=False,lw=2,annot=True,cbar=False,annot_kws = {'size':20,"weight": "bold"})
bottom, top = ax.get_ylim() 
left, right = ax.get_xlim()

plt.yticks(rotation= 0,fontsize = 30)
plt.setp(ax.get_xticklabels(), fontsize=16,fontweight='bold')
plt.setp(ax.get_yticklabels(), fontsize=16,fontweight='bold')
ax.set_title('Correlation Matrix',fontsize=40,fontweight='bold')
plt.show()
```

```python
X = df_feat
y = df['stroke']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# K-Nearest Neighbour Classifirer

```python
error_rate= []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))

plt.figure(figsize = (20,7),dpi=100)
plt.plot(range(1,40),error_rate,color = 'blue',marker = 'o',markerfacecolor='red',markersize = 6)
plt.title('Error Rate vs K-neighbours',fontsize=30)
plt.xlabel('K-neighbours',fontsize=25)
plt.ylabel('Error Rate',fontsize=25)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors = 9)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
pred
knn_cm = confusion_matrix(y_test,pred)
sns.heatmap(knn_cm,cmap='RdPu_r',vmin=-10,vmax=160,square=False,annot=True,cbar=False,fmt='0.0f',annot_kws = {'size':10,"weight": "bold"} )
print(accuracy_score(y_test, pred))
```

# Random Forest Classifier

```python
forest= RandomForestClassifier(n_estimators =40, random_state = 0)
forest.fit(X_train,y_train)  
y_pred = forest.predict(X_test)

rf_cm = confusion_matrix(y_test,y_pred)

sns.heatmap(rf_cm,cmap='RdPu_r',vmin=-10,vmax=160,square=False,annot=True,cbar=False,fmt='0.0f',annot_kws = {'size':10,"weight": "bold"} )

print(accuracy_score(y_test, y_pred))
```
